```python
# Import libraries
import vegan_functions as vf
import json
import os
from os.path import exists

# Store file check
file_exists = exists("storage.json")

if file_exists == False:
    store_dict = {} # store dictionary initialization
    gross = []      # gross profit
    net = []        # net profit

elif file_exists == True:
    with open("storage.json") as storage_js:
        json_dict = json.load(storage_js) 
        storage_js.close()
    store_dict = json_dict["storage"] # last store dictionary 
    gross = list(json_dict["gross"])  # last gross profit
    net = list(json_dict["net"])      # last net profit
    
else:
    print("ERROR: file error")


# While loop until closing program
while True:
    
    try:
        intention = str(input("\nInserisci un comando: ")) # Intention 

        if intention == "aggiungi":
            vf.vegan_add(store_dict)

        elif intention == "aiuto":
            vf.vegan_help()

        elif intention == "chiudi":
            vf.vegan_close() # Bye bye
            dict_test = {"storage": store_dict, "gross":gross, "net":net}
            dict_test_str = json.dumps(dict_test)
            
            with open("storage.json", "w") as output_file:
                json.dump(dict_test, output_file, indent = 6)
                output_file.close()
                
            break # Close program

        elif intention == "elenca":
            vf.vegan_list(store_dict)

        elif intention == "vendita":
            vf.vegan_sell(store_dict, gross, net)
            store_dict = {k: v for k, v in store_dict.items() if v[0] != 0} # Updating store in case of seeling

        elif intention == "profitti":
            vf.vegan_profit(gross, net)

        else:
            vf.vegan_help_mess()
            
    except:
        print("PROGRAMM ERROR")
```
